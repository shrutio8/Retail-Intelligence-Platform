AI BASED INSIGHT ENGINE 


```python
region_profit = df.groupby('region')['profit'].sum()

avg_profit = region_profit.mean()

for region, profit in region_profit.items():
    if profit < 0:
        print(f"{region} region is incurring losses and needs immediate attention.")
    elif profit < avg_profit:
        print(f"{region} region is underperforming compared to other regions.")
    else:
        print(f"{region} region is performing well with strong profits.")
```

```python
df.groupby(['region','category'])[['sales','profit']].sum()
```

```python
category_profit = df.groupby('category')['profit'].sum()

avg_profit = category_profit.mean()

for category, profit in category_profit.items():
    if profit < 0:
        print(f"{category} category is loss-making and needs immediate attention.")
    elif profit < avg_profit:
        print(f"{category} category is underperforming compared to others.")
    else:
        print(f"{category} category is performing well.")
```

```python
# Dynamic Category Insight
category_profit = df.groupby('category')['profit'].sum()
worst_category = category_profit.idxmin()

print(f"{worst_category} category is the least profitable and needs attention.")

# Dynamic Region Insight
region_profit = df.groupby('region')['profit'].sum()
worst_region = region_profit.idxmin()

print(f"{worst_region} region is underperforming compared to others.")

# Discount Impact
corr = df['discount'].corr(df['profit'])

if corr < -0.5:
    print("There is a strong negative relationship between discount and profit, indicating that higher discounts are significantly reducing profitability.")
elif corr < -0.2:
    print("Discounts have a moderate negative impact on profitability.")
else:
    print("Discounts do not have a significant impact on profitability.")
```
